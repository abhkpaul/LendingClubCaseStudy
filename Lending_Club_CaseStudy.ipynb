# Data Cleaning

**Checking for empty rows**

```python
loan_master.isnull().all(axis=1).sum()
```

**Checking nulls for each column**

```python
loan_master_nulls = loan_master.isnull().sum()
print(loan_master_nulls)
loan_master  = (loan_master[(loan_master_nulls > 0 == True)])
loan_master
loan_master_nulls = loan_master.isnull().sum()
loan_master_nulls = (loan_master_nulls[loan_master_nulls > 0])
loan_master_nulls

# loan_master = loan_master.dropna(axis=1, how='all')
# loan_master
```

**Creating a list of column names who have nulls for all its rows**

```python
null_counts_df = pd.DataFrame({'Column_Name': loan_master_nulls.index, 'Null_Count': loan_master_nulls.values})
null_counts_df = null_counts_df[(null_counts_df.Null_Count == total_rows)]
loan_master_nulls_list = list(null_counts_df.Column_Name.values)
loan_master_nulls_list
```

**Dropping All Columns who have nulls in all its rows**

```python
loan_master = loan_master.drop(columns=loan_master_nulls_list) 
loan_master
# loan_master_null_cleaned = loan_master.drop(columns=loan_master_nulls_list) 
# loan_master_null_cleaned
```

**Dropping All Columns which only have a single value for all its rows**

```python
columns_with_one_distinct_value = []

for column in loan_master.columns:
    single_value_column = loan_master[column].nunique()
    if single_value_column == 1:
        columns_with_one_distinct_value.append(column)

print(columns_with_one_distinct_value)

loan_master = loan_master.drop(columns=columns_with_one_distinct_value)
loan_master.head()
# columns_with_one_distinct_value = []

# for column in loan_master_null_cleaned.columns:
#     single_value_column = loan_master_null_cleaned[column].nunique()
#     if single_value_column == 1:
#         columns_with_one_distinct_value.append(column)

# print(columns_with_one_distinct_value)

# loan_master_singles_dropped = loan_master_null_cleaned.drop(columns=columns_with_one_distinct_value)
# loan_master_singles_dropped.head()
```

**Derive Year and Month column from Load Issue Date (issue_d)**

```python
loan_master.issue_d = pd.to_datetime(loan_master.issue_d, format='%b-%y')
loan_master['issue_y']=loan_master['issue_d'].dt.year
loan_master['issue_m']=loan_master['issue_d'].dt.month

loan_master.head(1)
```

**Standardizing 'emp_length' column with only Numbers**

```python
loan_master.emp_length.fillna('0',inplace=True)
loan_master.emp_length = pd.to_numeric(loan_master.emp_length.str.replace(r'\D', ''), errors='coerce')
loan_master.head()
```

**Standardizing 'term' column with only Numbers**

```python
# loan_master.term = loan_master.term.str.extract(r'(\d+)')
loan_master.term = pd.to_numeric(loan_master.term.str.replace(r'\D', ''), errors='coerce')
loan_master.head()
```

**Standardizing 'int_rate' and 'revol_util' columns by removing percentage symbols**

```python
loan_master.int_rate = pd.to_numeric(loan_master.int_rate.str.replace('%', ''), errors='coerce')
loan_master.revol_util = pd.to_numeric(loan_master.revol_util.str.replace('%', ''), errors='coerce')
loan_master
```

```python
distinct_counts = loan_master.nunique()
distinct_counts = distinct_counts[distinct_counts == total_rows].index.tolist()
distinct_counts
```

**Standardizing rest of the columns having numbers to numeric type (pd.to_numeric)**

```python
cast_cols = ['loan_amnt','funded_amnt','funded_amnt_inv','installment','annual_inc','dti','emp_length','total_pymnt']
load_master[cast_cols] = loan_master[cast_cols].apply(pd.to_numeric)
loan_master
```

**Dropping other unnecessary columns identified during Data understanding**

```python
misc_cols = ['url','desc','zip_code','addr_state']
loan_master = loan_master.drop(columns=misc_cols) 
loan_master
```

**Bivariate Analysis**

**Required Column Derivation**

```python
# Derived columns
# Creating Loan amount categorise into buckets.
loan['loan_amount_category'] = pd.cut(loan['loan_amnt'], [0, 7000, 14000, 21000, 28000, 35000], 
                                      labels=['0-7000', '7000-14000', '14000-21000', '21000-28000', '28000 +'])

# Creating annual incomes categories into buckets.
loan['annual_income_category'] = pd.cut(loan['annual_inc'], [0, 20000, 40000, 60000, 80000,1000000], 
                                        labels=['0-20000', '20000-40000', '40000-60000', '60000-80000', '80000 +'])

# Creating intrest rates categories into buckets.
loan['interest_rate_category'] = pd.cut(loan['int_rate'], [0, 10, 12.5, 16, 20], 
                                        labels=['0-10', '10-13', '12.5-16', '16 +'])
```

**Bivariate Analysis for Annual Income vs Chargeoff Proportion**

```python
out = loan.groupby(['annual_income_category', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
out['total'] = out['Charged Off'] + out['Current'] + out['Fully Paid'] 
out['chargeoff_proportion'] = out['Charged Off'] / out['total']
out.sort_values('chargeoff_proportion', ascending=False)
print(out)
fig, ax1 = plt.subplots(figsize=(12, 8),facecolor='w')
ax1.set_title('Annual Income vs Chargeoff Proportion',fontsize=15,color = 'b')
ax1=sns.barplot(x='annual_income_category', y='chargeoff_proportion', data=out)
ax1.set_ylabel('Chargeoff Proportion',fontsize=14,color = 'b')
ax1.set_xlabel('Annual Income',fontsize=14,color='b')
plt.show()

# Observation:
# Chargeoff Proportion is decreasing with annual income inscrease
# Income range 80000+  has less chances of getting charged off.
# Income range 0-20000 has high chances of getting charged off.
```

**Bivariate Analysis for Interest Rate vs Chargeoff Proportion**

```python
out = loan.groupby(['interest_rate_category', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
out['total'] = out['Charged Off'] + out['Current'] + out['Fully Paid'] 
out['chargeoff_proportion'] = out['Charged Off'] / out['total']
out.sort_values('chargeoff_proportion', ascending=False)
print(out)
fig, ax1 = plt.subplots(figsize=(12, 8),facecolor='w')
ax1.set_title('Interest Rate vs Chargedoff Proportion',fontsize=15,color='b')
ax1=sns.barplot(x='interest_rate_category', y='chargeoff_proportion', data=out)
ax1.set_xlabel('Interest Rate',fontsize=14,color='b')
ax1.set_ylabel('Chargeoff Proportion',fontsize=14,color = 'b')
plt.show()

# Observation:
# Charged off proportion is increasing with higher intrest rates.
# Interest of rate more than 16% has good chnaces of charged off as compared to other category intrest rates.
# interest of rate less than 10% has less chances of charged off.
```
